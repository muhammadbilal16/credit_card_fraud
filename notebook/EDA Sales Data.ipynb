# Customers Segmentation

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Create a new DataFrame with customer ID and total transaction amount
customer_df = df.groupby('transaction_id')['total'].sum().reset_index()

# Standardize the data
scaler = StandardScaler()
X = scaler.fit_transform(customer_df[['total']])

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Fit the K-means model with the optimal number of clusters
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
kmeans.fit(X)

# Add the cluster labels to the DataFrame
customer_df['cluster'] = kmeans.labels_

# Visualize the clusters
plt.scatter(X[:, 0], np.zeros_like(X), c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], np.zeros_like(kmeans.cluster_centers_), s=100, c='red')
plt.xlabel('Total transaction amount')
plt.title('Customer Segmentation')
plt.show()

# Analyze the characteristics of each cluster
customer_df.groupby('cluster')['total'].describe()

```

```python

# Select features for clustering
X = df[['quantity', 'unit_price', 'total']]

# Normalize the features
X = (X - X.mean()) / X.std()

# Perform KMeans clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)

# Add cluster labels to the original data
df['cluster'] = kmeans.labels_

# Calculate mean values of each feature by cluster
cluster_summary = df.groupby('cluster')[['quantity', 'unit_price', 'total']].agg(['count', 'mean', 'std', 'min', 'max'])

# Print the summary
print(cluster_summary)
```

```python
# Group the data by cluster and category
cluster_category_sales = df.groupby(['cluster', 'category']).agg({'total': 'sum'})

# Pivot the table to have clusters as rows and categories as columns
cluster_category_sales = cluster_category_sales.pivot_table(index='cluster', columns='category', values='total')

# Visualize the results
sns.heatmap(cluster_category_sales, cmap='YlGnBu', annot=True, fmt='.2f')
plt.title('Total Sales by Cluster and Category')
plt.xlabel('Category')
plt.ylabel('Cluster')
plt.show()

# Find the cluster with the highest sales for each category
for category in df['category'].unique():
    max_cluster = cluster_category_sales[category].idxmax()
    print(f'Customers in Cluster {max_cluster} tend to purchase more products in the {category} category')
```

# Findings from This Dataset

**The transaction amounts are right-skewed, indicating that there may be a few high-value transactions that are driving up the mean and standard deviation.** <br> <br>
**The most common customer type is basic, followed by gold and premium. Basic customers tend to make smaller purchases on average compared to gold and premium customers.** <br> <br>
**E-wallet is the most common payment type, followed by debit card and credit card. There may be some variation in payment type usage over time.** <br> <br>
**There is a positive correlation between quantity and total, indicating that larger quantities tend to lead to higher total sales. There is also a weak positive correlation between unit price and total, suggesting that higher-priced items may be associated with higher total sales.<br> Finally, there is a weak negative correlation between customer type and unit price, indicating that premium customers may purchase items with lower unit prices compared to basic or gold customers.** <br> <br>
**Sales appear to peak in the late morning to early afternoon, with the highest sales occurring between 11am and 2pm. Sales are generally lower in the early morning and late evening hours.** <br>


**The top-selling categories are Office Supplies, Technology, and Furniture, in that order.**

**The average unit price for Furniture is higher than the other two categories, while Technology has the lowest average unit price.**

**There are seasonal trends in sales, with higher sales in the second and fourth quarters.**

**The customer segmentation analysis identified four distinct segments based on their transaction history, with different spending patterns and preferences.**

**Cluster 3 has the highest average transaction amount and is the most profitable segment for the company.**

**Customers in Cluster 2 tend to purchase more products in the Furniture category, while customers in Cluster 1 tend to purchase more products in the Office Supplies category.**

**Targeting Cluster 3 with promotions and special offers can potentially increase sales and profitability for the company**
