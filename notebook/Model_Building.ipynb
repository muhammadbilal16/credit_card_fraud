```python
# Importing Libraries

import pandas as pd
from datetime import datetime
import sklearn
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importing Datasets

sales = pd.read_csv('data/sample_sales_data.csv')
sales.head()
```

```python
stock = pd.read_csv('data/sensor_stock_levels.csv')
stock.head()
```

```python
storage = pd.read_csv('data/sensor_storage_temperature.csv')
storage.head()
```

```python
# Check dataset info

sales.info()
```

```python
stock.info()
```

```python
storage.info()
```

```python
# Date data format conversion

sales['timestamp'] = pd.to_datetime(sales['timestamp'], format='%Y-%m-%d %H:%M:%S')
stock['timestamp'] = pd.to_datetime(stock['timestamp'], format='%Y-%m-%d %H:%M:%S')
storage['timestamp'] = pd.to_datetime(storage['timestamp'], format='%Y-%m-%d %H:%M:%S')
```

```python
# Function to covert timestamps to hourly

def hourly(df_name, col_name):
    
    timelist = []
    data = df_name[col_name].tolist()
    
    for each in data:
        each = each.strftime('%Y-%m-%d %H:00:00')
        timelist.append(each)
        
    df_name['timestamp'] = timelist
```

```python
hourly(sales, 'timestamp')
sales.head(3)
```

```python
hourly(stock, 'timestamp')
stock.head(3)
```

```python
hourly(storage, 'timestamp')
storage.head(3)
```

```python
# Save aggregate values for timestamps

sales_agg = sales.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum'}).reset_index()
sales_agg.head(3)
```

```python
stock_agg = stock.groupby(['timestamp', 'product_id']).agg({'estimated_stock_pct': 'mean'}).reset_index()
stock_agg.head(3)
```

```python
temp_agg = storage.groupby(['timestamp']).agg({'temperature': 'mean'}).reset_index()
temp_agg.head(3)
```

```python
# Merge datasets together

merged = stock_agg.merge(sales_agg, on=['timestamp', 'product_id'], how='left')
merged.head()
```

```python

merged = merged.merge(temp_agg, on='timestamp', how='left')
merged.head()
```

```python
# Finding columns with null values

merged.isnull().sum()
```

```python
# Filling in missing values

merged['quantity'] = merged['quantity'].fillna(merged['quantity'].mean())
merged.isnull().sum()
```

```python
# Adding category and unit price columns

prod_catg = sales[['product_id', 'category']]
prod_price = sales[['product_id', 'unit_price']]
```

```python
# Cheking for duplicates

print('Category -', prod_catg.duplicated().sum())
print('Price -', prod_price.duplicated().sum())
```

```python
# Deleting duplicates

prod_catg = prod_catg.drop_duplicates()
prod_price = prod_price.drop_duplicates()
```

```python
# Cheking for duplicates

print('Category -', prod_catg.duplicated().sum())
print('Price -', prod_price.duplicated().sum())
```

```python
# Merging into main dataset

merged = merged.merge(prod_catg, on="product_id", how="left")
merged = merged.merge(prod_price, on="product_id", how="left")
merged.head()
```

```python
merged.info()
```

```python
# Converting timestamp format

merged['timestamp'] = pd.to_datetime(merged.timestamp, format='%Y-%m-%d %H:%M:%S')
```

```python
# Timestamp extraction

merged['day_of_month'] = merged['timestamp'].dt.day
merged['day_of_week'] = merged['timestamp'].dt.dayofweek
merged['hour'] = merged['timestamp'].dt.hour
merged.drop(columns=['timestamp'], inplace=True)
merged.head()
```

```python
# Convert categories to numeric variables

merged = pd.get_dummies(merged, columns=['category'])
merged.head()
```

```python
merged.info()
```

```python
# Removing product id, no longer needed

merged.drop(columns='product_id', inplace=True)
merged.head(1)
```

```python
# Creating target variable y and independant variable x

X = merged.drop(columns=['estimated_stock_pct'])
y = merged['estimated_stock_pct']
print(X.shape)
print(y.shape)
```

```python
# Setting folds K
K = 10

# Setting split % to 75% training and 25% teasting
split = 0.75
```

```python
# Importing functions

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
```

```python
accuracy = []

for fold in range(0, K):

  # Instantiate algorithm
  model_ra = RandomForestRegressor()
  scaler = StandardScaler()

  # Create training and test samples
  X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=split, random_state=42)

  # Scale X data, we scale the data because it helps the algorithm to converge
  # and helps the algorithm to not be greedy with large values
  scaler.fit(X_train)
  X_train = scaler.transform(X_train)
  X_test = scaler.transform(X_test)

  # Train model
  trained_model = model_ra.fit(X_train, y_train)

  # Generate predictions on test sample
  y_pred = trained_model.predict(X_test)

  # Compute accuracy, using mean absolute error
  mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
  accuracy.append(mae)
  print(f"Fold {fold + 1}: MAE = {mae:.3f}")

print(f"Average MAE: {(sum(accuracy) / len(accuracy)):.2f}")
```

```python
# Fit the decision tree model
model = DecisionTreeRegressor(random_state=42)
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate the mean absolute error
mae = mean_absolute_error(y_test, y_pred)
print('MAE:', mae)
```

```python
model.score(X_train, y_train)
```

```python
model_ra.score(X_train, y_train)
```

```python
# Plotting importance of features

features = [i.split("__")[0] for i in X.columns]
importances = model_ra.feature_importances_
indices = np.argsort(importances)

fig, ax = plt.subplots(figsize=(10, 20))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
